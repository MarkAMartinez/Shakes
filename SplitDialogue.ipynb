```python
filename = "/Users/markmartinez/Downloads/Shakes-master/preprocessed_shakespeare.txt"
castfile = "/Users/markmartinez/Downloads/Shakes-master/curated_cast.txt"
#removes punctuation

remove = dict.fromkeys(map(ord, '\n ' + string.punctuation))
#remove space from this and newline
del remove[32]
del remove[10]

with open(filename,'r') as f:
   strippedFile = f.read().translate(remove).lower()

stops = stopwords.words("english")
stopdict = dict((s.lower(),None) for s in stops) 



castdict = {}
for line in open(castfile):
    words = nltk.word_tokenize(line.strip())
    for word in words:
        if(word not in stopdict):
            castdict[word.lower()] = True
```
